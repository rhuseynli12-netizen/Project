```python
import math
m=math.pow(4.2013*math.sqrt(0.1),1/3)+2-(1/3)*math.exp(2)+math.exp(-2)
r=math.sin(1/2*math.tan(-3/4)*math.log(5))
if m<2*r :
  print(math.sqrt(abs(3*m-5*r)))
if m>=2*r :
  print(math.sqrt(abs(3*m+5*r)))                                #13
```
